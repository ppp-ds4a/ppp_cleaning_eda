```python
import pandas as pd
import seaborn as sns
```

```python
PATH = "../data/webscrape_aggregated.csv"
PATH2 = "../data/under150k.csv"
PATH3 = "../data/even_and_odd.csv"

df_scrape = pd.read_csv(PATH3, index_col="Unnamed: 0")
df_150 = pd.read_csv(PATH2)

print(df_scrape.shape, df_150.shape)
```

```python
df_150 = df_150[df_150["Zip"].isnull() == False]
```

```python
stat_dict = {
    "LoanAmount": "sum",
    "JobsReported": "sum",
    "State": "count"
            }

subset_150 = df_150.groupby("Zip")[["LoanAmount", "JobsReported", "State"]].agg(stat_dict).reset_index()
subset_150.columns = ["zip", "loan_amount", "jobs_reported", "loan_count"]
subset_150.head()
```

```python
df_scrape = df_scrape.merge(subset_150, how="left", left_on="postal_code", right_on="zip").drop("zip", axis=1)
```

```python
df_scrape
```

```python
PATH3 = "ZIP-COUNTY-FIPS_2018-03.csv"

fips = pd.read_csv(PATH3)
```

```python
fips.head()
```

```python
fips_dict = dict(zip(fips["ZIP"], fips["STCOUNTYFP"]))

df_scrape["fips_code"] = df_scrape["postal_code"].map(fips_dict)
```

```python
df_scrape.to_csv("yelp_ppp_merge.csv")
```
