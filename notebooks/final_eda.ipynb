```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
PATH = "../data/yelp_ppp_merge.csv"
df= pd.read_csv(PATH)
df.shape
```

```python
df["cost_per_job"] = df["loan_amount"]/df["jobs_reported"]
```

```python
plt.style.use("seaborn-whitegrid")
df_bar2 = df[["postal_code","open_precovid", "open_postcovid", "loan_amount"]]
df_bar2.columns = ["Postal Code", "N-Businesses Open Pre-COVID", "N-Businesses Open January 2021", "Total Loan Amount"]
df_bar2\
.set_index("Postal Code")\
.sort_values("Total Loan Amount")\
.drop("Total Loan Amount", axis=1)\
.plot(
    kind="bar",
    figsize=(14,8),
    color=["powderblue", "thistle"],
    alpha=0.6,
)

plt.gca().xaxis.grid(False)
plt.xlabel("Postal Code")
plt.title("PPP Loans Helping Keep Businesses Open", fontsize="x-large")
plt.tight_layout()
plt.savefig("ppp_char3.png", dpi=150)
```

```python
plt.figure(figsize=(14,8))
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
sns.set_style("whitegrid")

g = sns.barplot(
    
    data = plot_df.reset_index(),
    x = "postal_code",
    y= "jobs_reported",
    color="powderblue",
    order = plot_df.reset_index().sort_values("jobs_reported")["postal_code"],
    alpha=0.6
)

plt.xticks(rotation = 90)

# Used for creating twin y axis chart to preserve scales
g2 = g.twinx()

g2.scatter(list(range(18)),
         plot_df_line['loan_amount'].values,
         linestyle='-',
         marker='o', 
         linewidth=1.0,
         c="salmon"
        )

# Loop for auto populating text labels
text_ls = ["$19m", "14.6m", "39.4m", "22.7m", "19.9m", "35.4m", "23.6m", "24.7m", "18.5m", "14.9m", "20.7m", "35.8m", "21.1m", "24.8m", "83.3m", "35.8m", "16.6m", "49m"]
for x, y in list(zip(list(range(18)), plot_df_line["loan_amount"])):
    
    plt.text(
        x=x, 
        y=y+500000, 
        s=f"${str(y)[:2]}.{str(y)[3]}m",
        fontsize="x-small",
        fontfamily="serif",
        rotation=50
    )
    
g2.grid(False)
g2.set(yticklabels=[])
g2.tick_params(right=False)

# Remove spines
g.spines["top"].set_visible(False)
g2.spines["top"].set_visible(False)
g.spines["right"].set_visible(False)
g2.spines["right"].set_visible(False)
g.spines["left"].set_visible(False)
g2.spines["left"].set_visible(False)

sns.set(font_scale=1.7)
g.set(xlabel="Postal Code", ylabel="Amount of Jobs Saved")
plt.title("Figure 1:The Cost to Keep Businesses Open", fontsize="x-large")
plt.legend(["Total <150k Loan Amount Received"], loc="upper left")
plt.tight_layout()
plt.savefig("ppp_char.png", dpi=150);
```
